### LC1: Given an array of integers `nums` and an integer `target`, return indices of the two numbers such that they add up to target.

### Solution 1

```c++
#include <iostream>
#include <vector>
#include <map>

std::vector<int> twoSum(){
    std::vector<int> nums = {};
    int target = {};
    int complement = {};
    std::map<std::vector, int> hashmap = {};

    for (int num : nums) {

        complement = target - nums[num];

        if (hashmap.count(complement)) {

            return {num, hashmap.count(complement)};

        }

        hashmap.count(nums[num]) = num;
    }

    return {};

}
```
